```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
y=iris.target
print(iris.DESCR)
print("targets---")
print(y)
print("features")
print(iris.feature_names)
```

```python
x=iris.data
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.55,random_state=65)
print(y_test)
```

```python
sl_rg=LinearRegression()
sl_rg.fit(x_train,y_train)
y_predict=sl_rg.predict(x_test)
print("---------TEST CLASS-------\n",y_test)
print("---------PREDICT CLASS-------\n",y_predict)
```

```python
mse=mean_squared_error(y_test,y_predict)
print("Mean squared error:",mse)
print(r2_score(y_test,y_predict))
```

```python
print(sl_rg.coef_)
print(sl_rg.intercept_)
```
